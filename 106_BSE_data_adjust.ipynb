## Historical stock price adjustment
Corporate actions are events initiated by a publicly traded company that can impact its stock price or the ownership structure of the company. These actions can lead to adjustments in historical stock prices to reflect the impact of the event accurately. Some common corporate actions that may require adjustments to historical stock prices include:

* **Dividends**: When a company declares and pays dividends to its shareholders, the stock price typically decreases by the amount of the dividend. This adjustment is made to reflect the fact that shareholders have received cash or additional shares.

* **Stock Splits**: A stock split occurs when a company divides its existing shares into multiple shares. For example, a 2-for-1 stock split doubles the number of shares while halving the price. Adjustments are made to historical stock prices to reflect the increased number of shares and the decreased share price after the split.

* **Stock Consolidations/Reverse Splits**: A stock consolidation, or reverse stock split, is the opposite of a stock split. It involves reducing the number of outstanding shares, resulting in a higher share price. Historical stock prices are adjusted to reflect the reduced number of shares and the increased share price.

* **Bonus Issues**: In a bonus issue, also known as a stock dividend, additional shares are issued to existing shareholders without any cash consideration. Historical stock prices are adjusted to account for the increased number of shares after the bonus issue.

* **Rights Issues**: In a rights issue, existing shareholders are given the right to purchase additional shares at a discounted price. Historical stock prices may need adjustment to reflect the dilution caused by the issuance of new shares.

* **Mergers and Acquisitions**: When two companies merge or one company acquires another, the stock prices of the involved companies may need adjustment to reflect the exchange ratio or cash consideration for the transaction.

These are just a few examples of corporate actions that can impact historical stock prices. Adjustments are made to ensure that historical data accurately reflects the impact of these events on the stock's performance and shareholder value.



```python
import time
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Example 1 : Stock Splits

**1.Read data from CSV into DataFrame**

```python
df = pd.read_csv('data/TATAsteel.csv')
df1=df[['Date', 'Open Price', 'High Price', 'Low Price', 'Close Price','No.of Shares']].copy()
df1
```

**2.Visualize data**

```python
plt.plot(np.flip(df1['Date']), np.flip(df1['Close Price']))
plt.title('TATA STEEL Stock price')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Price in Rs.')
```

!['Screenshot of split'](img/tata.jpg)

**3.Identify the data to be adjusted**

```python
split_date = pd.to_datetime('2022-07-28')
df1['Split'] = pd.to_datetime(df1['Date']).apply(lambda x: 10 if x < split_date else 1)
df1
```

**4.Calculate adjusted values**

```python
df1['AdjustedOpen'] = df1['Open Price'] / df1['Split']
df1['AdjustedHigh'] = df1['High Price'] / df1['Split']
df1['AdjustedLow'] = df1['Low Price'] / df1['Split']
df1['AdjustedClose'] = df1['Close Price'] / df1['Split']
df1['AdjustedVolume'] = df1['No.of Shares'] * df1['Split']
```

```python
df1
```

```python
fig, ax = plt.subplots()
ax.plot(np.flip(df1['Date']), np.flip(df1['AdjustedClose']),label= 'Adjusted close price')
# ax.plot(np.flip(df1['Date']), np.flip(df1['Close Price']),label= 'close price')
plt.title('TATA STEEL adjusted Stock price')
ax.legend()
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Price in Rs.')
```

## Example 2 : Bonus issues

```python
df = pd.read_csv('data/JETINFRA.csv')
df1=df[['Date', 'Open Price', 'High Price', 'Low Price', 'Close Price','No.of Shares']].copy()
df1
```

```python
plt.plot(np.flip(df1['Date']), np.flip(df1['Close Price']))
plt.title('JET INFRA Stock price')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Price in Rs.')
```

!['Screenshot of bonus'](img/jet.jpg)

```python
bonus_date = pd.to_datetime('2023-04-20')
df1['Bonus'] = pd.to_datetime(df1['Date']).apply(lambda x: 1 if x <= bonus_date else 0)
df1
```

```python
df1['AdjustedOpen'] = df1['Open Price'] / (df1['Bonus'] + 1)
df1['AdjustedHigh'] = df1['High Price'] / (df1['Bonus'] + 1)
df1['AdjustedLow'] = df1['Low Price'] / (df1['Bonus'] + 1)
df1['AdjustedClose'] = df1['Close Price'] / (df1['Bonus'] + 1)
df1['AdjustedVolume'] = df1['No.of Shares'] * (df1['Bonus'] + 1)
df1
```

```python
plt.plot(np.flip(df1['Date'].str[:2]), np.flip(df1['AdjustedClose']))
plt.title('JET INFRA adjusted Stock price')
plt.xlabel('Date')
plt.ylabel('Price in Rs.')
```
